```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#load csv files
df1 = pd.read_csv('files_for_lab/csv_files/file1.csv')
df2 = pd.read_csv('files_for_lab/csv_files/file2.csv')
df3 = pd.read_csv('files_for_lab/csv_files/file3.csv')
```

```python
#display dataframes for preliminary observation
display(df1.head())
display(df2.head())
display(df3.head())
```

```python
print(df2.shape)
df2.columns
```

```python
print(df3.shape)
df3.columns
```

```python
#rename columns before concatenation
col1 = [c.lower().replace(' ', '_') for c in df1.columns]
df1.columns = col1
df1 = df1.rename(columns={'st':'state'}).head()
df1.columns
```

```python
#remove percent sign from customer_lifetime_value.
'''I initially did this after concatenating all 3 csvs but decided to move the process before to facilitate
the division by 100 of values with the percent sign. This is under the assumption that these values were transformed
into percent.'''
df1.customer_lifetime_value = df1.customer_lifetime_value.replace('%','', regex = True).astype(float)/100
df1.info()
df1.head()
```

```python
col2 = [c.lower().replace(' ', '_') for c in df2.columns]
df2.columns = col2
df2 = df2.rename(columns={'st':'state'}).head()
df2.columns
```

```python
#repeat remove percent sign from customer_lifetime_value and division by 100 as with df1
df2.customer_lifetime_value = df2.customer_lifetime_value.replace('%','', regex = True).astype(float)/100
df2.info()
df2.head()
```

```python
col3 = [c.lower().replace(' ', '_') for c in df3.columns]
df3.columns = col3
df3.info()
df3.head()
```

```python
#concatenate 3 dataframes
a=pd.concat([df1,df2,df3])
a.head()
```

```python
#rearrange columns
b = a[['customer','gender','education','income','state','policy_type',
       'customer_lifetime_value','vehicle_class','monthly_premium_auto', 
       'number_of_open_complaints', 'total_claim_amount']]
```

```python
b.info()
b.head()
```

```python
b.describe()
```

```python
b.number_of_open_complaints.unique()
```

```python
# Delete the columns education and the number of open complaints from the dataframe
c = b.drop(['education','number_of_open_complaints'], axis = 1)
c.info()
```

```python
#clean variables
c.customer_lifetime_value = round(c.customer_lifetime_value,2)
c.customer_lifetime_value
c.total_claim_amount = round(c.total_claim_amount,2)
c.head(10)
```

```python
#check for missing values
c.isna().sum()
```

```python
#only 1 record with missing values so we can drop it
c.dropna()
```

## Categorical Variables (5)

- customer = id_number <br>
- gender = m/f or u for missing data <br>
- state = state where customer lives <br>
- policy type = insurance policy type <br>
- vehicle_class = class of vehicle



## Numeric Variables (4)

- income = annual income <br>
- customer_lifetime_value =  total income expected from customers for as long as they remain a client <br>
- monthly_premium_auto = monthly insurance payment by client
- total_claim_amount = amount paid by the insurance company as reimbursement against losses covered under the insurance policy

```python
# check for duplicates and drop if any
c.duplicated().value_counts()
c.drop_duplicates()
```

```python
# Filter out the data for customers who have an income of 0 or less
d = c[c.income > 0]
```
